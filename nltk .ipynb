```python
import nltk
nltk.download("all")
```

```python
file = open("nlp.txt", "r")
text = file.read()
print(text)
```

```python
from nltk.tokenize import sent_tokenize
nltk.tokenize.sent_tokenize(text,language='english')
sentences = sent_tokenize(text)

```

```python
print("Number of sentences:",len(sentences))
for i in range(len(sentences)):
    print("\nSentence ",i+1,":\n",sentences[i])
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(text)
```

```python
print("Total number of words: ",len(words))
print(words)
```

```python
words = word_tokenize(text,preserve_line=True)
len(words)
```

```python
words = word_tokenize(text)
len(words)
```

```python
from nltk.probability import FreqDist
all_fdist = FreqDist(words).most_common(20)
print(all_fdist)
```

```python

import matplotlib.pyplot as plt
import pandas as pd
all_fdist = pd.Series(dict(all_fdist))
fig, ax = plt.subplots(figsize=(7,7))
all_fdist.plot(kind='bar')
plt.title("Frequency Distribution of words")
plt.ylabel("Count")
```

```python
import re
text = re.sub('[^A-Za-z0-9]+', ' ', text)
```

```python
print(text)
```

```python
text = re.sub("\S*\d\S*", "", text).strip()
```

```python
print(text)
```

```python
import nltk
words = word_tokenize(text)
stopwords = nltk.corpus.stopwords.words('english')
words_sw_removed = []
for word in words:
    if word in stopwords:
        pass
    else:
        words_sw_removed.append(word)
```

```python
all_fdist = FreqDist(words_sw_removed).most_common(20)
all_fdist = pd.Series(dict(all_fdist))
fig, ax = plt.subplots(figsize=(7,7))

all_fdist.plot(kind='bar')
plt.title("Frequency Distribution of words")
plt.ylabel("Count")
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

stopwords = set(STOPWORDS)

wordcloud = WordCloud(width = 800, height = 800,
                    background_color ='white',
                    stopwords = stopwords,
                    min_font_size = 10).generate(text)

plt.figure(figsize = (7, 7), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
